```python
# get dimmies to gender
sex = pd.get_dummies(churn_data['gender'],drop_first=True)
```

```python
# contact dummies and main data set
churn_data= pd.concat([churn_data,sex,paymethod,contract], axis=1)
```

```python
# show data
churn_data
```

```python
churn_data.columns
```

```python
# drop unknow columns
churn_data.drop(['customerID','gender','PaymentMethod','Contract'],axis=1, inplace=True)
```

```python
# show data
churn_data
```

```python
# Encording the data set
lableEncoder = LabelEncoder()
churn_data_tf = churn_data.copy()
```

```python
# transform the data set
for i in churn_data.columns:
    churn_data_tf[i] = lableEncoder.fit_transform(churn_data[i])
```

```python
# show data
churn_data_tf 
```

```python
# get output and input features
X= churn_data_tf.drop(['Churn'],axis=1)
Y = churn_data_tf['Churn']
```

```python
# split x, y train and test
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, random_state=2)
```

```python
#  x, y train and test  shape
print(x_train.shape,'---> Training Feature Shape')
print(x_test.shape,'---> Testing Feature Shape')
print(y_train.shape,'---> Training Lable Shape')
print(y_test.shape,'---> Testing Lable Shape')
```

```python
# add LogisticRegression algoritham
churn_log_model = LogisticRegression()
```

```python
# fit data set
churn_log_model.fit(x_train,y_train)
```

```python
# get prediction
churn_prediction = churn_log_model.predict(x_test)
```

```python
# show prediction
churn_prediction
```

```python
# classification_report
print(classification_report(y_test,churn_prediction))
```

```python
# get accuracy score
accuracy_score_churn  = sm.accuracy_score(y_test,churn_prediction)
```

```python
# show accuracy score
print('Accuracy Score === ',str(accuracy_score_churn))
```

```python
mat = confusion_matrix(y_test, churn_prediction)
```

```python
# headt map test and prediction
sns.heatmap(mat.T,square=True,annot=True,fmt='d',cbar=False)
plt.xlabel('Accual Class')
plt.ylabel('Predicted Cass')
plt.show()
```
